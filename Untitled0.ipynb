```
#6
x = []
def fibona(n):
    k = int(n - 2)
    print("0 1", end=" ")
    a = 0
    b = 1
    i = 1

    while i <= k:
        x.append(a + b)
        print(a + b, end=" ")
        i = i + 1
        h = b
        b = a + b
        a = h

fibona(1000)
print(x)
```

```
import math

def nguyen_to(n):

    if (n < 2):
        return False;

    # check so nguyen to khi n >= 2
    squareRoot = int(math.sqrt(n));
    for i in range(2, squareRoot + 1):
        if (n % i == 0):
            return False;
    return True;
x = []

x.append(2)
for i in range(3, 10000 + 1):
    if (nguyen_to(i)):
        x.append(int(str(i)))
    i = i + 2;
print(x);
```
